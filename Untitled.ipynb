<h4>Линейная коррекция</h4>

```python
from skimage.io import imread, imsave, imshow
from skimage import img_as_float
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = imread('audi.jpg')
imshow(img)
img_f = img_as_float(img)
```

```python
img_temp = (img_f - img_f.min()) * 255.0 / (img_f.max() - img_f.min())
result = img_temp.astype('uint8')
imshow(result)
```

<h4>Модель серого мира</h4>

```python
img = imread('zoloto.jpg')
imshow(img)
```

```python
img_f = img_as_float(img)
r = img_f[:,:,0]
g = img_f[:,:,1]
b = img_f[:,:,2]
avg = (r / 2 + g / 2 + b /2) / 3
rw = (r / 2) / avg
gw = (g / 2) / avg
bw = (b / 2) / avg
np.seterr(invalid='ignore')
red = r / rw 
imshow(red)
```

```python
green = g / gw
imshow(green)
```

```python
blue = b / bw
imshow(blue)
```

```python
rgb = np.dstack((red, green, blue))
imshow(rgb)
```

<h4>Автоконтраст цветного изображения</h4>

```python
img = cv.imread('kontrast.jpg')
plt.imshow(img)
```

```python
img_trnsf = cv.cvtColor(img, cv.COLOR_RGB2YUV)
img_trnsf[:,:,0] = cv2.equalizeHist(img_trnsf[:,:,0])
result = cv.cvtColor(img_trnsf, cv.COLOR_YUV2BGR)
plt.imshow(result)
```

<h4>Выравнивание Гистограммы</h4>

```python
img = cv.imread('prikol.jpg', 0)
fig = plt.figure(figsize=(7,7))
fig.add_subplot(2,2,1)
plt.imshow(img, cmap="gray")

histo, bins = np.histogram(img.flatten(), 256, [0,256])

cdf = histo.cumsum()
cdf_normalized = cdf*float(histo.max())/cdf.max()

fig.add_subplot(2,2,2)
plt.plot(cdf_normalized, color = 'b')
plt.hist(img.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.show()
```

```python
fig2 = plt.figure(figsize=(7,7))
cdf = (cdf-cdf[0]) *255/ (cdf[-1]-1)
cdf = cdf.astype(np.uint8)
img2 = cdf[img]
fig2.add_subplot(2,2,1)
plt.plot(cdf_normalized, color = 'b')
plt.hist(img2.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.show()
fig2.add_subplot(2,2,2)
plt.imshow(img2, cmap="gray")
```
